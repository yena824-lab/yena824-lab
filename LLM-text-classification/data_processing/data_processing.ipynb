```python
train_data['소분류'] = train_data['원인요소'].astype(str).apply(lambda x: x.split('>')[-1] if pd.notna(x) else '')
val_data['소분류'] = val_data['원인요소'].astype(str).apply(lambda x: x.split('>')[-1] if pd.notna(x) else '')
test_data['소분류'] = test_data['원인요소'].astype(str).apply(lambda x: x.split('>')[-1] if pd.notna(x) else '')
```

```python
train_data["대분류"] = train_data["원인요소"].astype(str).apply(lambda x: x.split(">")[0] if pd.notna(x) else "")
train_data["중분류"] = train_data["원인요소"].astype(str).apply(lambda x: x.split(">")[1] if pd.notna(x) and len(x.split(">")) > 1 else "")
val_data["대분류"] = val_data["원인요소"].astype(str).apply(lambda x: x.split(">")[0] if pd.notna(x) else "")
val_data["중분류"] = val_data["원인요소"].astype(str).apply(lambda x: x.split(">")[1] if pd.notna(x) and len(x.split(">")) > 1 else "")
test_data["대분류"] = test_data["원인요소"].astype(str).apply(lambda x: x.split(">")[0] if pd.notna(x) else "")
test_data["중분류"] = test_data["원인요소"].astype(str).apply(lambda x: x.split(">")[1] if pd.notna(x) and len(x.split(">")) > 1 else "")
```

```python
import re
import pandas as pd

# 첫 번째 괄호 삭제 대상
first_bracket_remove = {
    '금속', '벌레', '이물질일반', '플라스틱', '곰팡이독소일반', '기생충일반',
    '동물질병', '식물질병', '결핵균', '곰팡이수', '노로바이러스', '대장균',
    '리스테리아 모노사이토제네스', '미생물일반', '병원성 대장균',
    '신규식품', '안전', '위생', '건강', '나트륨', '열량', '영양성분일반',
    '난류', '밀', '알레르기일반', '우유', '원료·성분·함량', '제품명',
    '해당없음', '기구용기포장유래물질일반', '우루시올',
    '동물용의약품일반', '식품첨가물일반', '의약품성분일반', '잔류농약일반',
    '유리'
}

# 두 번째 괄호 삭제 대상
second_bracket_remove = {
    '살모넬라', '쌀(GMO)', '지질', '질소화합물', '탄수화물',
    '기간', '호모실데나필', '말라치온', '납'
}

def clean_last_part(value):
    if pd.isna(value):
        return ''
    
    # 마지막 요소 추출
    last_part = value.split('>')[-1]
    
    # "기타"로 시작하면 기타 처리 우선
    if last_part.startswith('기타'):
        return re.sub(r'기타.*', '기타', last_part)
    
    # 길이가 3단계 이상이면 bracket 규칙 적용
    parts = value.split('>')
    if len(parts) >= 3:
        third_value = parts[2].split('(')[0].strip()
        
        if third_value in first_bracket_remove:
            parts[2] = re.sub(r'\(.*$', '', parts[2])
        elif third_value in second_bracket_remove:
            parts[2] = re.sub(r'\)\(.*$', ')', parts[2])
    
    return parts[-1]

# '원인요소' → '소분류' 생성
train_data['소분류'] = train_data['원인요소'].astype(str).apply(clean_last_part)
val_data['소분류'] = val_data['원인요소'].astype(str).apply(clean_last_part)
test_data['소분류'] = test_data['원인요소'].astype(str).apply(clean_last_part)
```

```python
replace_map = {
    'GMO(LMO)일반(기타)': 'GMO(LMO)일반',
    'GMO(LMO)일반(연어(GMO))': 'GMO(LMO)일반',
    '쌀(GMO)(할랄)': '쌀(GMO)'
}

train_data['소분류'] = train_data['소분류'].replace(replace_map)
val_data['소분류'] = val_data['소분류'].replace(replace_map)
test_data['소분류'] = test_data['소분류'].replace(replace_map)
```

```python
train_left = set(row_data['소분류'].dropna().unique())
train_right = set(train_data['소분류'].dropna().unique())

train_common = train_left & train_right
train_left_only = train_left - train_right
train_right_only = train_right - train_left

print(f"공통 조합 수: {len(train_common)}")
print(f"식품안전분류표에만 있는 조합 수: {len(train_left_only)}")
print(f"url 데이터에만 있는 조합 수: {len(train_right_only)}")
```

```python
val_left = set(row_data['소분류'].dropna().unique())
val_right = set(val_data['소분류'].dropna().unique())

val_common = val_left & val_right
val_left_only = val_left - val_right
val_right_only = val_right - val_left

print(f"공통 조합 수: {len(val_common)}")
print(f"식품안전분류표에만 있는 조합 수: {len(val_left_only)}")
print(f"url 데이터에만 있는 조합 수: {len(val_right_only)}")
```

```python
test_left = set(row_data['소분류'].dropna().unique())
test_right = set(test_data['소분류'].dropna().unique())

test_common = test_left & test_right
test_left_only = test_left - test_right
test_right_only = test_right - test_left

print(f"공통 조합 수: {len(test_common)}")
print(f"식품안전분류표에만 있는 조합 수: {len(test_left_only)}")
print(f"url 데이터에만 있는 조합 수: {len(test_right_only)}")
```
